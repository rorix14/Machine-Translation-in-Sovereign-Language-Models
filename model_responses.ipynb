# Make Models EU-PT translation dataset

```python
import numpy as np
import random
from datasets import load_from_disk, Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer
from tqdm import tqdm
import torch
```

```python
seed = 42
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
random.seed(seed)
np.random.seed(seed)

device = torch.cuda.current_device()
```

```python
def translation_base_dataset(dataset_path):
    en_por_dataset = load_from_disk(dataset_path)

    en_inputs_formated = ["English: " + sample["en"] + " Portuguese:" for sample in en_por_dataset]
    return en_inputs_formated
```

```python
def translation_chat_dataset(dataset_path):
    en_por_dataset = load_from_disk(dataset_path)

    en_inputs_chat_formated = []
    for sample in tqdm(en_por_dataset):
        message = [
            {"role": "system",
             "content": "You are a translation model. Translate English to European Portuguese. Return ONLY the translation."},
            {"role": "user",
             "content": "Translate: " + sample["en"]},
        ]

        en_inputs_chat_formated.append(message)

    return en_inputs_chat_formated
```

```python
def base_translate(model_path, dataset_path, output_dir_path, gpu_run=True):
    model = AutoModelForCausalLM.from_pretrained(model_path)
    if gpu_run:
        model.to(0)

    tokenizer = AutoTokenizer.from_pretrained(model_path)
    tokenizer.pad_token = tokenizer.eos_token

    model_responses = []
    max_tokens = model.config.max_position_embeddings - 296

    translation_dataset = translation_base_dataset(dataset_path)

    for sample in tqdm(translation_dataset):
        if gpu_run:
            inputs = tokenizer(sample, return_tensors="pt").to(0)
        else:
            inputs = tokenizer(sample, return_tensors="pt")

        with torch.no_grad():
            outputs = model.generate(**inputs, max_new_tokens=max_tokens)

        response = outputs[:, inputs["input_ids"].shape[-1]:][0]
        model_responses.append(tokenizer.decode(response, skip_special_tokens=True))

    results_data = [{"pt_generated": out} for out in model_responses]
    results_dataset = Dataset.from_list(results_data)
    results_dataset.save_to_disk(output_dir_path)
```

```python
def chat_translate(model_path, dataset_path, output_dir_path, gpu_run=True):
    model = AutoModelForCausalLM.from_pretrained(model_path)
    if gpu_run:
        model.to(0)

    tokenizer = AutoTokenizer.from_pretrained(model_path)

    model_responses = []
    max_tokens = model.config.max_position_embeddings - 296

    translation_dataset = translation_chat_dataset(dataset_path)

    for sample in tqdm(translation_dataset):
        if gpu_run:
            input_chat = tokenizer.apply_chat_template(sample, return_tensors="pt", add_generation_prompt=True).to(0)
        else:
            input_chat = tokenizer.apply_chat_template(sample, return_tensors="pt", add_generation_prompt=True)

        with torch.no_grad():
            output = model.generate(input_chat, max_new_tokens=max_tokens, pad_token_id=tokenizer.eos_token_id)

        response = output[0][input_chat.shape[-1]:]
        model_responses.append(tokenizer.decode(response, skip_special_tokens=True))

    results_data = [{"pt_generated": out} for out in model_responses]
    results_dataset = Dataset.from_list(results_data)
    results_dataset.save_to_disk(output_dir_path)
```

```python
dataset_path = "datasets/frmt/frmt_en_pt_translation_dataset"  # Example dataset
model_base_path = "utter-project/EuroLLM-1.7B"  # Example base model
model_instruct_path = "utter-project/EuroLLM-1.7B-Instruct"  # Example instruct model
```

Translation prompts for base EuroLLM models

```python
base_translate(model_base_path, dataset_path, "datasets/frmt data/frmt_eurollm_1_7_responses")
```

Translation prompts for instruct models

```python
chat_translate(model_instruct_path, dataset_path, "datasets/frmt data/frmt_eurollm_1_7_Instruct_responses")
```
