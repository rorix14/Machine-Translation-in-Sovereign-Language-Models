# Evaluate models EU-PT translation dataset

```python
import numpy as np
import pandas as pd
import random
from datasets import load_from_disk, concatenate_datasets, Dataset
from transformers import pipeline
from evaluate import load
from comet import download_model, load_from_checkpoint
from evaluate.visualization import radar_plot
import torch
import seaborn as sns
```

```python
seed = 42
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
random.seed(seed)
np.random.seed(seed)

device = 0 if torch.cuda.is_available() else "cpu"
torch.cuda.set_device(device)
```

```python
class Comet:
    def __init__(self, device="cpu"):
        comet_model_path = download_model("Unbabel/wmt22-comet-da")
        self.comet_model = load_from_checkpoint(comet_model_path)
        self.comet_model.to(device)

    def compute(self, predictions, references, source):
        comet_data = [
            {"src": s, "mt": m, "ref": r}
            for s, m, r in zip(source, predictions, references, strict=True)
        ]

        comet_model_output = self.comet_model.predict(comet_data, batch_size=16, gpus=1)
        return {"scores": comet_model_output["scores"], "score": comet_model_output["system_score"]}
```

```python
class VidPtEu:
    def __init__(self, references, device="cpu"):
        self._pipe_vid = pipeline(model="liaad/PtVId", device=device)

        self.ref_classification = self._pipe_vid(references)

        self.ref_length = len(self.ref_classification)

        self.ref_pt_eu_indexes, self.ref_pt_eu_count, self.ref_pt_eu_confidence_average = self._count_pt_eu(
            self.ref_classification)

    def compute(self, predictions):
        assert len(
            predictions) == self.ref_length, "Length of predictions dataset does not match references dataset length!"

        predictions_classification = self._pipe_vid(predictions)

        pre_pt_eu_indexes, pre_pt_eu_count, pre_pt_eu_confidence_average = self._count_pt_eu(predictions_classification)

        return {"indexes": pre_pt_eu_indexes, "total": pre_pt_eu_count, "confidence": pre_pt_eu_confidence_average,
                "score": pre_pt_eu_count / self.ref_pt_eu_count, "raw info": predictions_classification}

    def _count_pt_eu(self, dataset):
        pt_eu_indexes = []
        pt_eu_count = 0
        pt_eu_confidence_average = 0

        for i, sample in zip(range(0, self.ref_length), dataset):

            if "PT-PT" in sample["label"]:
                pt_eu_indexes.append(i)
                pt_eu_count = pt_eu_count + 1
                pt_eu_confidence_average = pt_eu_confidence_average + sample["score"]

        return pt_eu_indexes, pt_eu_count / self.ref_length, pt_eu_confidence_average / pt_eu_count
```

```python
def pt_eu_metrics(references, predictions_list, sources, device="cpu"):
    metric_sacrebleu = load("sacrebleu")
    metric_rouge = load('rouge')
    metric_comet = Comet(device)
    metric_vid = VidPtEu(references=references, device=device)

    metrics_data = []
    for predictions in predictions_list:
        results_sacrebleu = metric_sacrebleu.compute(predictions=predictions, references=references)
        results_rouge = metric_rouge.compute(predictions=predictions, references=references)
        results_comet = metric_comet.compute(predictions=predictions, references=references, source=sources)
        results_vid = metric_vid.compute(predictions=predictions)

        metrics_data.append({"sacrebleu": results_sacrebleu["score"], "rougeL": results_rouge["rougeL"],
                             "comet": results_comet["score"], "vid": results_vid["score"]})

    return metrics_data
```

```python
def frmt_get_model_test_data(model_data):
    frmt_test_predictions = []
    temp_list = model_data[entity_len:  entity_len + len(entity_test_sentences)]
    frmt_test_predictions.extend(temp_list)

    start_len_test = entity_len + len(entity_test_sentences) + lexical_len
    temp_list = model_data[start_len_test:  start_len_test + len(lexical_test_sentences)]
    frmt_test_predictions.extend(temp_list)

    start_len_test = entity_len + len(entity_test_sentences) + lexical_len + len(lexical_test_sentences) + random_len
    temp_list = model_data[start_len_test:  start_len_test + len(ramdom_test_sentences)]
    frmt_test_predictions.extend(temp_list)

    return frmt_test_predictions
```
